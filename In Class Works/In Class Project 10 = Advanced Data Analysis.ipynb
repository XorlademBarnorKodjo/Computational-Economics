### plot data by selected countries from the data

```python
pairs = [("CAN", "FIN"), ("FRA", "DEU"), ("USA", "GBR", "ESP", "MEX")]
linestyles = ["-", ":", "--", "-."]
```

```python
for pair in pairs:  
    fig, ax = plt.subplots(figsize=(16,8))  
    for i in range(len(pair)):  
        country = pair[i]  
        linestyle = linestyles[i]  
        data.loc[country,:]["cgdppc"].dropna(
        ).plot.line(ax = ax,  
               label = country, 
               linestyle = linestyle)
```

### Creating Indicator Variable

```python
def create_indicator_variable(data, 
                             indicator_name,
                             index_name,
                             target_index_list):
    #prepare column with name of indicator
    data[indicator_name] = 0
    # for each index whose name matches an entity in target_index_list a value of i will be recorded
    for index in target_index_list:
        #check if entity index (ISO code) of row matches the entity from target index list
        # set valuess of indicator name(coluumn) to 1 for all such rows
        data.loc[index, [indicator_name]] = int(1)
data = fraser_data

#select "ISO_Code" from namess of double index

index_name = data.index.names[0]
indicator_name = "North America"
# Cuba, Grenada, Saint Kittss, Saint Lucas, Saint Vincent are missing
countries_in_north_america = [
                            "BHS", "BRB", "BLZ", "CAN", "CRI", "DOM", "SLV",
                            "GTM", "HTI", "HND", "JAM", "MEX", "NIC", "PAN",
                            "TTO", "USA"]
create_indicator_variable(data = data,
                          indicator_name = "North America",
                          index_name = index_name,
                          target_index_list = countries_in_north_america)
```
