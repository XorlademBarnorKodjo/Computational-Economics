# Two Dimensional List and Arrays

## Boolean Indexing

Allows us to create a true or fals index that allows us to include or disinclude values from a list

```python
import random

rand_list = [random.random() * 10 for i in range(10)]
rand_list
```

Helps us to select a range or subset of a data we are dealing with example: data from countries with GDP below 1000 or only female respondents

```python
subset = [val for val in rand_list if val >3]
subset
```

```python
subset = [random.random() * 10 for i in range(10) if i > 3]
print(subset)
```

# Create Random Data for DataFrame

```python
macro_dict = {"GDP": {},
            "Money": {},
            "Real GDP": {},
            "Price Level": {},
             "Population": {}}
```

```python
for key in macro_dict:
    for i in range(1990, 2018):
        macro_dict[key][i] = np.random.random() * 10000
```

```python
macro_dict
```

```python
#convert dictionary to dataframe

macro_df = pd.DataFrame(macro_dict)
macro_df
```

```python
#adding new variables which are computed from existing ones

macro_df["Velocity"] = macro_df["GDP"].div(macro_df["Money"])
macro_df["GDPperCapita"] = macro_df["GDP"].div(macro_df["Population"])
macro_df
```

```python
#calling rows from a year

macro_df.loc[2000]
```

```python
#calling rows from a range of years

macro_df.loc[2000:2005]
```

```python
#calling only some variables

macro_df[["Money", "GDP"]]
```

```python
#calling rows from a range of years for particular variables

macro_df.loc[2000:2005, ["Money", "GDP"]]
```

```python
#plotting the entire dataframe

import matplotlib.pyplot as plt
%matplotlib inline
macro_df.plot.line()
```

```python
for key in macro_df:
    macro_df[key].plot.line()
    plt.title(key, fontsize = 22)
    plt.show()
    plt.close()
```

```python
#fig line allows various features

for key in macro_df:
    fig, ax = plt.subplots(figsize = (12, 8))
    macro_df[key].plot.line()
    plt.title(key, fontsize = 22)
    plt.show()
    plt.close()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

a = 2  # number of rows
b = 3  # number of columns
c = 1  # initialize plot counter

fig = plt.figure(figsize=(14,10))

for key in macro_df:
    plt.subplot(a, b, c)
    plt.title(key)
    sns.lineplot(data = macro_df[key])
    c = c + 1
    plt.tight_layout()

    
plt.show()
```
